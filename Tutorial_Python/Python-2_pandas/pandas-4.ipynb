### 2-5. Conditional selection

```python
df.Sex == 'female'
```

```python
df.loc[df.Sex == 'female']
```

```python
df[df['Sex'] == 'female']
```

```python
df.loc[:,'Age']
```

#### Two conditions

```python
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
```

```python
df.loc[(df.Sex == 'female') & (df.Age < 30)]
```

```python
df.loc[(df.Sex == 'female') | (df.Age < 30)]
```

```python
reviews.loc[reviews.country.isin(['Italy', 'France'])]
```

```python
df.loc[df.Age.isin([19, 20])]
```

```python
df[(df['Age'] == 19) | (df['Age'] == 20)]
```

```python
[df.loc[df.Age.isin([19, 20])].shape, df[(df['Age'] == 19) | (df['Age'] == 20)].shape]
```
